```python
def sigmoid(z):
    
    denominator = np.add(1, np.exp(-z))
    
    sig = np.divide(1, denominator)
    
    return sig
```

```python
def compute_loss(y, y_hat):
    
    diff = np.subtract(y, y_hat)
    
    diff_square = np.square(diff)
    
    loss = np.divide(diff_square, -2)
    
    return loss
```
